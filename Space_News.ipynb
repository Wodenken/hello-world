You should submit a notebook that contains a function that can be run that retrieves the space.com news page, extracts the news stories, and prints out the headline, author, synopsis, and date and time for each story.

```
from time import time
from time import sleep 
from random import randint 
from IPython.core.display import clear_output
from warnings import warn 
from bs4 import BeautifulSoup
import requests 
```

```
url = 'https://www.space.com/news'
response = requests.get(url)
if(response.ok):
  data = response.text
  soup = BeautifulSoup(data, 'html.parser')
  raw_articles = soup.find_all(class_='content')
  print(raw_articles)
```

```
raw_articles = soup.select('content')
for article in raw_articles:
  headline = article.select_one('h3').get_text()
  author = article.select_one('span > span').get_text()
  synopsis = article.select_one('div > p').get_text()
  date_time = article.select_one('time')
  article = {'headline': headline, 'author': author, 'synopsis': synopsis, 'date_time': date_time}
  articles.append(article)
print(articles)
```

```
def space_news(soup, articles):
  raw_articles = soup.select('content')
  for article in raw_articles:
    headline = article.select_one('h3').get_text()
    author = article.select_one('span > span').get_text()
    synopsis = article.select_one('div > p').get_text()
    date_time = article.select_one('time')
    article = {'headline': headline, 'author': author, 'synopsis': synopsis, 'date_time': date_time}
    articles.append(article)
start_time = time()
request_count = 0
articles = []
has_next_page = True
MAX_REQUESTS = 70
page_number = 1
query = {'sort':'i', 'pg': page_number}
url = 'https://www.space.com/news'
headers = {'user-agent':'articlescraper - school project (wodenk@gmail.com)'}
while has_next_page and request_count <= MAX_REQUESTS:
  clear_output(wait = True)
  response = requests.get(url, params=query, headers=headers)
  if(response.ok):
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    space_news(soup, articles)
    next_button = soup.select('.listings-pagination-button-next')
    has_next_page = len(next_button) > 0
  else:
    warn('Launch Sequence #: {}, Rejected - status code: {}'.format(request_count, response.status_code))
  request_count += 1
  sleep(randint(1, 3))
  elapsed_time = time() - start_time
  print('Requests: {}, Frequency: {} requests/s, {} articles processed.'.format(request_count, request_count/elapsed_time, len(articles)))
  page_number += 1
print('Scraping complete.')
print('Requests: {}, Frequency: {} requests/s, {} articles processed.'.format(request_count, request_count/elapsed_time, len(articles)))
```
