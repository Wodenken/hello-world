4) Top-10 Highest Pledged

```
bools = (df['category'] == 'Documentary') & (df['state'] == 'successful') 
s_docs = df.iloc[bools.values]
s_docs_topten = s_docs.nlargest(10, columns='pledged')
s_docs_topten
```

5) Average_per_backer:

-The rows with 0 backers returned "NaN".

-This may be remedied with the "fillna()" or "dropna()" function. 

```
backers_df = df.assign(average_per_backer = (df.pledged) / (df.backers))
backers_df.head(10)
```

6) DataFrame without "0 backers". 

```
existing_backers_df = backers_df.dropna(axis=0, how='any')
existing_backers_df.head(10)
```

-----------> ***nycflights13***

1) Merging DataFrames: "flights" & "airlines"

```
flights_durty = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/flights.csv')
flights = flights_durty.dropna(axis=0, how='any')
```

```
airlines = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/airlines.csv')
```

```
nycflights13 = pd.merge(flights, airlines)
nycflights13.head(13)
```
